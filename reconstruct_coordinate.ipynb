```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from glob import glob
```

```python
import json
file_path = ".\\data\\raw\\shoe\\transforms.json"
with open(file_path, "r") as file:
    json_data = json.load(file)

    camera_angle_x = json_data["camera_angle_x"]
    frames = json_data["frames"]

X = []
Y = []
Z = []
T = []
for frame in frames:
    img_path = frame["file_path"]
    transform_matrix = frame["transform_matrix"]
    
    x = transform_matrix[0][:3]
    y = transform_matrix[1][:3]
    z = transform_matrix[2][:3]
    t = [transform_matrix[0][3], transform_matrix[1][3], transform_matrix[2][3]]

    X.append(x)
    Y.append(y)
    Z.append(z)
    T.append(t)
```

```python
# Tạo figure và axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Thêm tọa độ điểm gốc
ax.scatter(0, 0, 0, c='red')  # Điểm thứ nhất, màu đỏ
ax.quiver(0, 0, 0, 1, 0, 0, length=1,color='red') # x
ax.quiver(0, 0, 0, 0, 1, 0, length=1,color='green') # y
ax.quiver(0, 0, 0, 0, 0, 1, length=1,color='blue') # z

for i in range(len(X)):
    ax.scatter(T[i][0], T[i][1], T[i][2], c='blue')  # Điểm thứ hai, màu xanh
    if i == 1:
        ax.quiver(T[i][0], T[i][1], T[i][2], X[i][0], X[i][1], X[i][2], length=1, color='red')
        ax.quiver(T[i][0], T[i][1], T[i][2], Y[i][0], Y[i][1], Y[i][2], length=1, color='green')
        ax.quiver(T[i][0], T[i][1], T[i][2], Z[i][0], Z[i][1], Z[i][2], length=1, color='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
